```python
#ANSWER-1
import matplotlib.pyplot as p
import numpy as np

sig1 = [5* np.sin(2*np.pi*15*t) for t in np.linspace(0,1,1000)]
sig2 = [5* np.cos(2*np.pi*2*t) for t in np.linspace(0,1,1000)]
sig3 = [5* np.sin(2*np.pi*2000*t) for t in np.linspace(0,1,1000)]
fsig=[]

for i,j,k in zip(sig1,sig2,sig3):
    fsig.append(i+j+k)
    


ft = np.fft.fft(fsig)
p.plot(ft)

p.plot(ft[:len(ft)//2])
p.show()
```

```python
x = np.array([11, 5, 5, 18, 20, 9])
len(x)       
```

```python
N = (len(x)-1)/2 
n = np.array([i for i in np.arange(-N, N+1)]) 
N,n
```

```python
L = []         
l = []
tim = [i for i in np.linspace(-N, N, 1000)] 
n = np.arange(-N, N+1)
for i in n:   
    for t in tim:
        term = 1
        for k in np.arange(-N, N+1):
            if k!=i:
                term = term*((t-k)/(i-k))   
        l.append(term)        
    L.append(l)     
    l=[]

    
for i in range(len(L)):
    p.plot(tim, L[i])
p.show() 
```

```python
newL = [i*np.array(j) for i,j in zip(x, L)]
y_=sum(newL)

p.stem(n, y, 'black')
p.plot(tim, y_,'red')
p.show()

kp = float(input('Enter x: '))
sp = 0

for xi,yi in zip(n, x):
    sp += yi * np.prod((kp - n[n != xi])/(xi - n[n != xi]))
    
print('For x = ', kp, ', y = ', sp)
```
